```python
import re
```

# search()
매칭되는 문자를 찾아서 
True/False 로 return

```python
hand = open('mbox-short.txt')

for line in hand:
    line = line.rstrip() #각 열의 마지막 개행문자 제거
    if re.search('From:', line):  #line.find() 와 동일한 결과
        print line
```

```python
hand = open('mbox-short.txt')

for line in hand:
    line = line.rstrip() #각 열의 마지막 개행문자 제거
    if re.search('^From:.+@', line):  #line.find() 와 동일한 결과
        print line
```

```python
# findall()
# 입력한 정규식에 매칭되는 문자를 찾아내서 return (list)
```

```python
#greedy
str = "From: kkoseul@gmail.com, skkim@mail.com"
print re.findall("^From: (.+@\S+)",str)
```

```python
#non-greedy
str = "From: kkoseul@gmail.com, skkim@mail.com"
print re.findall("^From: (.+?@\S+)",str)
```

```python
s = 'Hello from csev@umich.edu to cwen@iupui.edu about the metting @2PM'
lst = re.findall('\S+@\S+',s)
print lst
# \S 정규표현식을 @ 좌우로 사용해서 '@2PM'은 반환되지 않음
```

```python
hand = open('mbox-short.txt')

for line in hand:
    line = line.rstrip() #각 열의 마지막 개행문자 제거
    x = re.findall('\S+@\S+',line)
    if len(x)>0:  #정규식과 일치하는 문자열이 없으면 비어있는 list를 return 합니다.
        print x
        
        #email 주소와 함께 있는 다른 문자들까지 다 나옴...
```

```python
hand = open('mbox-short.txt')

for line in hand:
    line = line.rstrip() 
    x = re.findall('[a-zA-Z0-9]\S*@\S*[a-zA-Z]',line)
    if len(x)>0:  
        print x
```

```python
hand = open('mbox-short.txt')

for line in hand:
    line = line.rstrip() 
    x = re.findall('\w+[\w\.]*@\w+[\w\.]*\.[A-Za-z]+',line) # \w = [A-Za-z0-9_]
    if len(x)>0:  
        print x
```

```python
hand = open('mbox-short.txt')

for line in hand:
    line = line.rstrip() 
    x = re.findall('^Details:.*rev=([0-9.]+)',line)
    if len(x)>0:  
        print x
```

```python
hand = open('mbox-short.txt')

for line in hand:
    line = line.rstrip() 
    x = re.findall('^From .* ([0-9][0-9]):',line) #From으로 시작하는 줄의 시간만 추출
    if len(x)>0:  
        print x
```

```python
#Ex 11.1
hand = open('mbox-short.txt')

reg = raw_input("Enter a regular expression: ")
row_cnt = 0

for line in hand:
    line = line.rstrip() 
    x = re.findall(reg,line) #From으로 시작하는 줄의 시간만 추출
    if len(x)>0:  
        row_cnt += 1
        
#print row_cnt
print 'mbox-short.txt had ',row_cnt,' lines that matched ',reg
```

```python
#Ex 11.2
file_name = raw_input("Enter file: ")
hand = open(file_name)
total = 0
row_cnt=0

for line in hand:
    line = line.rstrip() 
    x = re.findall('^New Revision: ([0-9.]+)',line) #From으로 시작하는 줄의 시간만 추출
    if len(x)>0:  
        total += int(x[0])
        row_cnt +=1
        
print total/row_cnt
```
