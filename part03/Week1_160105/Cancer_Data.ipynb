### 향후 데이터 분석 계획
- Protein Expression, DNA Methylation, Copy Number 를 개별로 암종류 예측 모델을 만들어 봄.
- 2가지 데이터씩을 묶어서 암종류 예측 모델 생성,  3가지 데이터씩을 묶어서 암종류 예측 모델 생성
- 4가지 이상은 데이터가 부족해서 불가능해보임.
- Images 데이터안에서 암조직 세포의 사진이 있으므로, 세포사진으로 어떤 암조직인가 예측 모델도 가능하지 않을까 생각됨.

###  데이터 분석시 문제점 
- 105개의 연구에서 분석된 데이터들이 표준적인 방법으로 이루어지지 않음.
- mRNA Expression 데이터를 생성한 플랫폼이 동일하지 않음.
- RNA-seq 데이터는 암종류별 샘플 조직이 다르고, 인종별, 연령별, 성별 특징을 어떻게 처리할지를 고민 필요.
     - 1000 genome 데이터를 표준으로 잡고 할 수도 있을것 같음
