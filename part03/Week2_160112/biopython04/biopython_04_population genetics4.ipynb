- 4) F-statistics 계산

We will analyze the 10 percent sampling of chromosome 2 without the offspring.**(hapmap10_auto_noofs_2.gp)**

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from Bio.PopGen.GenePop import Controller as gpc
%matplotlib inline
```

```python
my_pops = [l.rstrip() for l in open('hapmap10_auto_noofs_2.pops')]
num_pops = len(my_pops)
```

GenePop application과 interaction하는 controller 생성

```python
ctrl = gpc.GenePopController()
```

.calc_fst_all

- 각 인종에 대해 multilocus로 Fis, Fst, Fit 계산

    - Fis = 1-(Hi/Hs)
    - Fst = 1-(Hs/Ht)
    - Fit = 1-(Hi/Ht)
![Fst](http://www.reed.edu/biology/professors/srenn/fst.jpg)

```python
(multi_fis, multi_fst, multi_fit), f_iter = ctrl.calc_fst_all('hapmap10_auto_noofs_2.gp')
```

```python
print(multi_fis, multi_fst, multi_fit)
```

- 각 locus에 대한 fst, fis, fit을 담은 후 

```python
fst_vals = []
fis_vals = []
fit_vals = []
for f_case in f_iter:
    name, fis, fst, fit, qinter, qintra = f_case
    fst_vals.append(fst)
    fis_vals.append(fis)
    fit_vals.append(fit)
```

- Fst - histogram
- Fis, Fit는 violin plot

```python
sns.set_style("whitegrid")
fig = plt.figure(figsize=(16, 9))
ax = fig.add_subplot(1, 1, 1)
ax.hist(fst_vals, 50, color='r')
ax.set_title('FST, FIS and FIS distributions')
ax.set_xlabel('FST')
ax = fig.add_subplot(2, 2, 2)
sns.violinplot(data=[fis_vals, fit_vals], ax=ax, vert=False)
ax.set_yticklabels(['FIS', 'FIT'])
ax.set_xlim(-.1, 0.4)
```

.calc_fst_pair
- **avg** : 모든 인종 사이의 average pair-wise Fst 계산
- **fpair**

```python
fpair_iter, avg = ctrl.calc_fst_pair('hapmap10_auto_noofs_2.gp')
```

```python
min_pair = min(avg.values())
max_pair = max(avg.values())
arr = np.ones((num_pops - 1, num_pops - 1, 3), dtype=float)
```

```python
sns.set_style("white")
fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(111)
for row in range(num_pops - 1):
    for col in range(row + 1, num_pops):
        val = avg[(col, row)]
        norm_val = (val - min_pair) / (max_pair - min_pair)
        ax.text(col - 1, row, '%.3f' % val, ha='center')
        if norm_val == 0.0:
            arr[row, col - 1, 0] = 1
            arr[row, col - 1, 1] = 1
            arr[row, col - 1, 2] = 0
        elif norm_val == 1.0:
            arr[row, col - 1, 0] = 1
            arr[row, col - 1, 1] = 0
            arr[row, col - 1, 2] = 1
        else:
            arr[row, col - 1, 0] = 1 - norm_val
            arr[row, col - 1, 1] = 1
            arr[row, col - 1, 2] = 1
ax.imshow(arr, interpolation='none')
ax.set_title('Multilocus Pairwise FST')
ax.set_xticks(range(num_pops - 1))
ax.set_xticklabels(my_pops[1:])
ax.set_yticks(range(num_pops - 1))
ax.set_yticklabels(my_pops[:-1])
```

Fst (between YRI and CEU)을 **all_fsts**에 담고, LCT(Lactase) gene 영역에서의 Fst를 **inside_fsts**에 담는다.

```python
pop_ceu = my_pops.index('CEU')
pop_yri = my_pops.index('YRI')
start_pos = 136261886  # b36
end_pos = 136350481
all_fsts = []
inside_fsts = []
for locus_pfst in fpair_iter:
    name = locus_pfst[0]
    pfst = locus_pfst[1]
    pos = int(name.split('/')[-1])  # dependent
    my_fst = pfst[(pop_yri, pop_ceu)]
    if my_fst == '-':  # Can be this
        continue
    all_fsts.append(my_fst)
    if pos >= start_pos and pos <= end_pos:
        inside_fsts.append(my_fst)
```

Fst median/ Fst mean/Fst 90% percentile/Fst 99% percentile

```python
print(inside_fsts)
print('%.2f/%.2f/%.2f/%.2f' % (np.median(all_fsts), np.mean(all_fsts), np.percentile(all_fsts, 90), np.percentile(all_fsts,99)))
```
